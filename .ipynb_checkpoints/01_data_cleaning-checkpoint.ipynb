```python
df = pd.read_csv('data/raw/archive/train.csv')
```

```python
df.head()
```

```python
print(f"Shape: {df.shape}")
print("\nColumns:\n", df.columns)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[^\w\s]', '', regex=True)
```

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Parse order_date as datetime
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', dayfirst=True)
```

```python


# Create directory if it doesn't exist
os.makedirs('../data/processed', exist_ok=True)

# Now save the cleaned data
df.to_csv('../data/processed/clean_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Make sure 'order_date' is datetime
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')

# Plot number of orders per day
df.set_index('order_date')['id'].resample('D').count().plot(figsize=(10,5))
plt.title('Orders Over Time')
plt.ylabel('Number of Orders')
plt.xlabel('Date')
plt.grid(True)
plt.show()
```

```python

# Remove non-numeric parts and convert to int
df['time_takenmin'] = df['time_takenmin'].astype(str).str.extract(r'(\d+)').astype(int)


# Now you can safely compute the mean and plot
df.groupby('city')['time_takenmin'].mean().sort_values().plot(kind='barh')
plt.title('Average Delivery Time by City')
plt.xlabel('Minutes')
plt.show()
```

```python
df['hour'] = df['order_date'].dt.hour
sns.countplot(x='hour', data=df)
plt.title('Orders by Hour of Day')
plt.show()
```
